```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def convert_to_zev(baseline, sales, mean_co2):
    
    allowances = sales * baseline
    
    activity = sales * mean_co2
    
    surplus = allowances - activity
    
    if surplus < 0:
        zev_credits = 0
    else:
        zev_credits = surplus / 167
    
    if zev_credits > (sales*.085):
        zev_credits = (sales*0.085)
    
    return zev_credits
```

```python
sales_volumes = list(range(0, 200000))
mean_co2 = 60
baseline = 160
zev_credits = [convert_to_zev(baseline=baseline, sales=x, mean_co2=mean_co2) for x in sales_volumes]
```

```python
df = pd.DataFrame(data = {
    'sales_volumes': sales_volumes,
    'zev_credits': zev_credits
})
```

```python
fig, ax = plt.subplots()

ax.plot(sales_volumes, zev_credits)

ax.set_xlim(0, 200000)
ax.set_ylim(0, 200000)
```

```python
df.plot(x='sales_volumes', y='zev_credits')
```
